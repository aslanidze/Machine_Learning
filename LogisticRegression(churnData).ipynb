```python
import pandas as pd
from scipy.stats import chi2_contingency
from sklearn.feature_selection import chi2

# بارگذاری داده‌ها
data = pd.read_csv("ChurnData.csv")

# جداسازی متغیر هدف و ویژگی‌ها
X = data.drop("churn", axis=1)  # ویژگی‌ها
y = data["churn"]  # متغیر هدف

# بررسی داده‌های عددی
numeric_features = X.select_dtypes(include=['float64', 'int64'])
categorical_features = X.select_dtypes(include=['object'])

# 1. آزمون chi-square برای داده‌های دسته‌بندی‌شده
for column in categorical_features:
    contingency_table = pd.crosstab(data[column], y)
    _, p, _, _ = chi2_contingency(contingency_table)
    print(f"Feature: {column}, p-value: {p}")

# 2. آزمون chi2 برای داده‌های عددی
from sklearn.feature_selection import SelectKBest, chi2
chi2_scores, p_values = chi2(numeric_features, y)

for i, column in enumerate(numeric_features.columns):
    print(f"Feature: {column}, p-value: {p_values[i]}")
```

```python
features_to_drop = ['tollmon', 'callwait', 'confer', 'logtoll', 'lninc', 'custcat', 'voice', 'pager']
X_cleaned = X.drop(columns=features_to_drop)
```

```python
from sklearn import preprocessing
X_sc = preprocessing.StandardScaler()

X_transfer = X_sc.fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X_transfer, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
yhat = LR.predict(X_test)
yhat
```

```python
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test, yhat,pos_label=0)
```
